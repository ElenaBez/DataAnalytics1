Poisson Distribution

```python
#If we randomly choose 8 employees in NWR, what is the probability of 5 of them beeing employees of Innovative industry?

from scipy.stats import poisson
# Expected number of employees working in innovative industries (lambda)
expected_count = 0.23 * 8  # 23% of 8 employees

# Calculate the probability of finding k employees in innovative industries
k = 5
probability = poisson.pmf(k, expected_count)

print(f"The probability of finding 5 employees working in innovative industries is approximately {probability:.4f}")
```

```python
# Create an array of possible values (0 to 8 employees)
k_values = np.arange(0, 8)

# Calculate the probabilities for each value of k
probabilities = [poisson.pmf(k, expected_count) for k in k_values]

# Create a bar chart to visualize the Poisson PMF
plt.bar(k_values, probabilities, align='center', alpha=0.8)
plt.xlabel('Number of Employees Working in Innovative Industries in NWR')
plt.ylabel('Probability')
plt.title('Poisson Probability  for 8 Employees')

plt.show()
```

Normal Distribution

```python
#Average number of employeer in any given industry and in any given region is 6695 in a year, standard diviation is 7787 ( see df description in line 21). What is the probability of employee number in industry to be 5000? 
mean = 6695
std_dev = 7787
x = 5000

# Calculate the probability using the CDF
probability = stats.norm.cdf(x, loc=mean, scale=std_dev)

print(f"The probability of the number of employees being less than or equal to {x} is: {probability:.4f}")
```

```python
# Generate a range of values from -5 standard deviations to +5 standard deviations
values = range(int(mean - 5 * std_dev), int(mean + 5 * std_dev))
probabilities = [stats.norm.cdf(value, loc=mean, scale=std_dev) for value in values]

# Plot the CDF
plt.plot(values, probabilities, marker='o')
plt.xlabel("Number of Employees")
plt.ylabel("CDF Probability")
plt.title("CDF of Employee Numbers in Regional Industry")
plt.grid(True)

# Highlight the point at x=10000
plt.axvline(x, color='red', linestyle='--', label=f'x = {x}')
plt.legend()

# Set the x-axis limits to start from 0
upper_limit = int(mean + 5 * std_dev)
values = range(0, upper_limit)
plt.xlim(0, upper_limit)
plt.show()
```

```python
#Average number of employeers in Ireland  in any given industry is 50087, standard diviation is 25020 ( see description in line 83). What is the probability of employee number in industry to be 70000? 
mean = 50087
std_dev = 25020
x = 70000
# Calculate the probability using the CDF
probability = stats.norm.cdf(x, loc=mean, scale=std_dev)

print(f"The probability of the number of employees being less than or equal to {x} is: {probability:.4f}")
```

```python
# Generate a range of values from -7 standard deviations to +7 standard deviations
values = range(int(mean - 7 * std_dev), int(mean + 7 * std_dev))
probabilities = [stats.norm.cdf(value, loc=mean, scale=std_dev) for value in values]

# Plot the CDF
plt.plot(values, probabilities, marker='o')
plt.xlabel("Number of Employees")
plt.ylabel("CDF Probability")
plt.title("CDF of Employee Numbers in Industry in Ireland")
plt.grid(True)

# Highlight the point at x=10000
plt.axvline(x, color='red', linestyle='--', label=f'x = {x}')
plt.legend()

# Set the x-axis limits to start from 0
upper_limit = int(mean + 7 * std_dev)
values = range(0, upper_limit)
plt.xlim(0, upper_limit)
plt.show()
```

```python
#Average percentage of employees working in innovative industries in NWR is 16%, standard diviation is 5% ( see description in line 83). What is the probability thet NWR innovative employment % is more or equal to the Rest of Ireland innovative employment percentage (33%)?, (see line 136)? 
mean = 16
std_dev = 5
x = 33
# Calculate the probability using the CDF
probability = 1-stats.norm.cdf(x, loc=mean, scale=std_dev)

print(f"The probability of the innovative industries employment in NWR being more than or equal to {x} which is innovative industries employment in the Rest of Ireland is: {probability:.4f}")
```

```python
# Generate a range of values from -5 standard deviations to +5 standard deviations
values = range(int(mean - 5 * std_dev), int(mean + 5 * std_dev))
probabilities = [1-stats.norm.cdf(value, loc=mean, scale=std_dev) for value in values]

# Plot the CDF
plt.plot(values, probabilities, marker='o')
plt.xlabel("Percentage")
plt.ylabel("CDF Probability")
plt.title("CDF of percentage of employees working in innovative industries in NWR being equal to the rest of Ireland ")
plt.grid(True)

# Highlight the point at x=10000
plt.axvline(x, color='red', linestyle='--', label=f'x = {x}')
plt.legend()

# Set the x-axis limits to start from 0
upper_limit = int(mean + 5 * std_dev)
values = range(0, upper_limit)
plt.xlim(0, upper_limit)
plt.show()
```
