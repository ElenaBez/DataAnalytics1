Poisson Distribution

```python
#If we randomly choose 8 employees in NWR, what is the probability of 5 of them being employees of Innovative industry?

from scipy.stats import poisson
# Expected number of employees working in innovative industries (lambda)
expected_count = 0.23 * 8  # 23% of 8 employees

# Calculate the probability of finding k employees in innovative industries
k = 5
probability = poisson.pmf(k, expected_count)

print(f"The probability of finding 5 employees working in innovative industries is approximately {probability:.4f}")
```

```python
# Create an array of possible values (0 to 8 employees)
k_values = np.arange(0, 8)

# Calculate the probabilities for each value of k
probabilities = [poisson.pmf(k, expected_count) for k in k_values]

# Create a bar chart to visualize the Poisson PMF
plt.bar(k_values, probabilities, align='center', alpha=0.8)
plt.xlabel('Number of Employees Working in Innovative Industries in NWR')
plt.ylabel('Probability')
plt.title('Poisson Probability  for 8 Employees')

plt.show()
```

Normal Distribution

```python
#Average number of employees in any given industry and in any given region is 6695 in a year, standard deviation is 7787 ( see df description in line 21). What is the probability of employee number in industry to be 5000? 
mean = 6695
std_dev = 7787
x = 5000

# Calculate the probability using the CDF
probability = stats.norm.cdf(x, loc=mean, scale=std_dev)

print(f"The probability of the number of employees being less than or equal to {x} is: {probability:.4f}")
```

```python
# Generate a range of values from -5 standard deviations to +5 standard deviations
values = range(int(mean - 5 * std_dev), int(mean + 5 * std_dev))
probabilities = [stats.norm.cdf(value, loc=mean, scale=std_dev) for value in values]

# Plot the CDF
plt.plot(values, probabilities, marker='o')
plt.xlabel("Number of Employees")
plt.ylabel("CDF Probability")
plt.title("CDF of Employee Numbers in Regional Industry")
plt.grid(True)

# Highlight the point at x=10000
plt.axvline(x, color='red', linestyle='--', label=f'x = {x}')
plt.legend()

# Set the x-axis limits to start from 0
upper_limit = int(mean + 5 * std_dev)
values = range(0, upper_limit)
plt.xlim(0, upper_limit)
plt.show()
```

```python
# Generate a range of values from -7 standard deviations to +7 standard deviations
values = range(int(mean - 7 * std_dev), int(mean + 7 * std_dev))
probabilities = [stats.norm.cdf(value, loc=mean, scale=std_dev) for value in values]

# Plot the CDF
plt.plot(values, probabilities, marker='o')
plt.xlabel("Number of Employees")
plt.ylabel("CDF Probability")
plt.title("CDF of Employee Numbers in Industry in Ireland")
plt.grid(True)

# Highlight the point at x=10000
plt.axvline(x, color='red', linestyle='--', label=f'x = {x}')
plt.legend()

# Set the x-axis limits to start from 0
upper_limit = int(mean + 7 * std_dev)
values = range(0, upper_limit)
plt.xlim(0, upper_limit)
plt.show()
```

```python
#Average percentage of employees working in innovative industries in NWR is 16%, standard diviation is 5% ( see description in line 41). What is the probability thet NWR innovative employment % is more or equal to the Rest of Ireland innovative employment percentage (33%)?, (see line 41)? 
mean = 16
std_dev = 5
x = 33
# Calculate the probability using the CDF
probability = 1-stats.norm.cdf(x, loc=mean, scale=std_dev)

print(f"The probability of the innovative industries employment in NWR being more than or equal to {x} which is innovative industries employment in the Rest of Ireland is: {probability:.4f}")
```

```python
data=filtered_df
```

```python

# Select the relevant columns for clustering
X = data[['Year', 'VALUE']]
```

```python
# Standardize the data (mean=0, std=1)
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
# Determine the optimal number of clusters using the Elbow method
wcss = []
for i in range(1, 11):
    kmeans = KMeans(n_clusters=i, init='k-means++', random_state=42)
    kmeans.fit(X_scaled)
    wcss.append(kmeans.inertia_)

plt.plot(range(1, 11), wcss)
plt.title('Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS (Within-Cluster Sum of Squares)')
plt.show()
```

```python

# Based on the Elbow method, choose an appropriate number of clusters
k = 4 

# Perform K-means clustering with the chosen number of clusters
kmeans = KMeans(n_clusters=k, init='k-means++', random_state=42)
y_kmeans = kmeans.fit_predict(X_scaled)

# Add the cluster labels to the original dataset
data['cluster'] = y_kmeans
# Define the custom tick locations and labels for the x-axis
custom_ticks = [2000, 2003, 2006, 2009, 2012, 2015, 2018, 2021]
custom_tick_labels = [str(Year) for Year in custom_ticks]

# Visualise the clusters
for i in range(k):
    cluster_data = data[data['cluster'] == i]
    plt.scatter(cluster_data['Year'], cluster_data['VALUE'], label=f'Cluster {i + 1}')

plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], s=200, c='red', label='Centroids')
plt.title('K-means Clustering')
plt.xlabel('Year')
plt.ylabel('Value')
plt.xticks(custom_ticks, custom_tick_labels)
plt.xlim(2000, max(data['Year'])) 
plt.legend(loc='center left', bbox_to_anchor=(1, 0.5))

plt.show()

# access the cluster assignments in the 'cluster' column of the 'data' DataFrame
print(data[['Year', 'Region', 'Sector', 'VALUE', 'cluster']])

```
