```python
#Create a dataframe combining sectors: Mordern manufacturing, Traditional manufacturing, Information & Communication technology.Also combine Border, West and Mid West Regions.  Also 
Sector_mapping={"Agriculture, forestry & fishing, mining & quarrying, food, beverages and tobacco ": "Traditional", "Traditional Manufacturing ": "Traditional", "Energy, Water, Waste & Construction ":"Traditional","Selected services ":"Traditional", "Information & Communication Services ":"Innovative", "Modern Manufacturing ":"Innovative"}
Region_mapping={"Border":"Northern and Western Region","West":"Northern and Western Region","Mid West":"Northern and Western Region","South East":"Rest of Ireland", "South West":"Rest of Ireland","Dublin":"Rest of Ireland","Mid East":"Rest of Ireland", "Midlands":"Rest of Ireland"}
df["Regions of Ireland"]=filtered_df["Region"].map(Region_mapping)
df["Industry"]=filtered_df["Sector"].map(Sector_mapping)
grouped_df=df.groupby(["Regions of Ireland", "Industry", "Year"])["VALUE"].sum().reset_index() 
print(grouped_df)
```

```python
grouped_df.describe()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sns.scatterplot(data=grouped_df, x='Year', y='VALUE', alpha=1, hue='Industry', size="Regions of Ireland")
plt.title('Innovative vs Traditional industries employment')
plt.legend(loc='upper right', bbox_to_anchor=(1.55, 1));
                                        
```

```python
# Group by Year, calculate the total value for each Year

regions_year_totals = grouped_df.groupby(['Regions of Ireland','Year'])['VALUE'].transform('sum')
grouped_df['Relative Percentage'] = (grouped_df['VALUE'] /regions_year_totals)*100
print(grouped_df)
```

```python
sns.scatterplot(data=grouped_df, x='Year', y='Relative Percentage', alpha=1, hue='Industry', size="Regions of Ireland")
plt.title('Innovative vs Traditional industries employment')
plt.legend(loc='upper right', bbox_to_anchor=(1.55, 1));
```

```python
northern_western_region_df = grouped_df[grouped_df['Regions of Ireland'] == 'Northern and Western Region']
summary_statistics = northern_western_region_df.describe()
print(summary_statistics)
```

```python
df = pd.DataFrame(grouped_df)

heatmap_data = df.pivot_table(index='Year', columns='Industry', values='Relative Percentage', aggfunc='first')

plt.figure(figsize=(16, 9))
heatmap = sns.heatmap(heatmap_data, cmap='coolwarm', annot=True, fmt='.0f', annot_kws={'size': 10})
plt.title('Regions of Ireland Heatmap')
plt.xlabel('Regions of Ireland')
plt.ylabel('Year')

cbar = heatmap.collections[0].colorbar

plt.show()
```
